## 3️⃣ Data Cleaning

```python
sns.heatmap(df.isnull()) # Making the heatmap to visual missing value in the data set
```

```python
df['Age'].isnull().mean() * 100 # finding the persentage of missing age data
```

```python
df['Age'] = df['Age'].fillna(df['Age']).mean() # filling the null values from mean 
```

```python
df['Cabin'].isnull().mean() * 100 # finding the persentage of  missing cabin data
```

```python
df['Cabin'] = df['Cabin'].fillna(df['Cabin']).mode()[1] # filling the null values from mode
```

```python
df['Embarked'].isnull().mean() * 100  # finding the persentage of  missing Embarked data
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked']).mode()[0] # filling the null values from mode
```

```python
df.isnull().sum()   # again checking the null value
```

```python
sns.heatmap(df.isnull()) # again Making the heatmap to visual missing value in the data set
```

## 4️⃣ Data Visualization (EDA)

```python
sns.countplot(x= 'Survived',data=df) 
plt.title('surviver data')
plt.show()
```

```python
sns.countplot(x = 'Survived',data=df , hue = 'Sex')
plt.title('surviver by gender')
plt.show()
```

```python
sns.countplot(x = 'Embarked',data=df , hue = 'Sex')
plt.title('Embarked by gender')
plt.show()
```

```python
sns.countplot(x = 'Survived',data=df , hue = 'Pclass')
plt.title('surviver by pclass')
plt.show()
```

```python
sns.barplot(x="Pclass", y="Fare", hue="Survived", data=df, ci=None)
plt.title("Average Fare by Class and Survival Status")
plt.show()
```

```python
sns.barplot(x='Cabin', y='Survived', data=df)
plt.xticks([0,1], ['No Cabin', 'Has Cabin'])
plt.ylabel('Survival Probability')
plt.show()
```

```python
survival_rate = df.groupby('Age')['Survived'].mean().reset_index() # common age survival rate
print(survival_rate)
```

```python
avg_age = df.groupby('Survived')['Age'].mean().reset_index() # common age survival rate by geneder
print(avg_age)
```

```python
avg_fare = df.groupby(['Pclass', 'Survived'])['Fare'].mean().reset_index() # average fare paid by class and survival status
print(avg_fare)
```

```python
survival_by_port = df.groupby('Embarked')['Survived'].mean().reset_index() # survival rates for passengers by embarkation port (Embarked)
print(survival_by_port)
```
